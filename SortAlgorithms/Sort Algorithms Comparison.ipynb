#### Random Array Generator

```python
def randomArrayGen(N, K = 20, maxValue = 20000, seed = 42):
    """Return an array with K arrays of size N with random values up to maxValue"""
    # we're going to fix the seed, in order to obtain the same array every time
    nprnd.seed(seed)
    return nprnd.randint(maxValue+1, size=(K, N)).tolist()
```

#### Benchmarking Method

```python
def compareAlgorithms(algorithms, expRange=(1, 14), timeAlgo = timeAlgo):
    """Apply each algorithms to each map and plot the results.
    
    Algorithms: List with the algorithms to be benchmarked
    expRange: all the arrays has a power of two element, 
            the exponent range is expressed by expRange
    """
    arrays = [randomArrayGen(2**x) for x in range (expRange[0], expRange[1] + 1)]
    for algorithm in algorithms:
        result = []
        for array in arrays:
            result.append(timeAlgo(algorithm, array))
        plt.plot(range(expRange[0], expRange[1] + 1), result, label = algorithm.__name__, animated=True)
    plt.legend(loc=2)
    plt.show()
```

```python
def timeAlgo(algorithm, arrays):
    """given an algorithm and an input array, measure time to process input.
    return, the average time between all the inputs"""
    algoName = algorithm.__name__
    strSetup = "from __main__ import " + algoName
    sample = []
    for array in arrays:
        strArray = str(array)
        command = algoName + "(" + strArray + ")"
        # repeat = num. Repetitions of algorithm; number = num. loops in the same input
        sample.append(repeat(command, setup = strSetup, repeat = 1, number = 1))
    return numpy.median(sample)
```

## Insertion Sort

An sorting algorithm which moves elements one at a time into the correct position. The algorithm consists of inserting one element at a time into the previously sorted part of the array, moving higher ranked elements up as necessary. [see more at Wikipedia](http://en.wikipedia.org/wiki/Insertion_sort)

```python
def insertionSort(array):
    """Return the array passed as argument sorted, using insertion sort."""
    # Check if array is empty
    if not array:
        return []
    else:
        for i in xrange(1, len(array)):
            j = i - 1
            key = array[i]
            while (array[j] > key) and (j >= 0):
                array[j+1] = array[j]
                j -= 1
            array[j+1] = key
        return array
```

## Counting Sort

A sorting technique based on keys between a specific range. It works by counting the number of objects having distinct key values (kind of hashing). Then doing some arithmetic to calculate the position of each object in the output sequence. [see more at Wikipedia](http://en.wikipedia.org/wiki/Counting_sort)

```python
# We use defaultdict to avoid having to initialize an array with zeros.
from collections import defaultdict

def countingSort(array):
    """Return the array passed as argument sorted, using counting sort."""
    # Check if array is empty
    if not array:
        return []
    else:
        countArray = defaultdict(int)
        result = []
        for element in array:
            countArray[element] += 1
        for i in xrange(1, len(array) + 1):
            result += [i] * countArray[i]
        return result
```

## Quicksort

Quicksort is a divide and conquer algorithm. Quicksort first divides a large array into two smaller sub-arrays: the low elements and the high elements. Quicksort can then recursively sort the sub-arrays.

The steps are:

1. Pick an element, called a pivot, from the array.
2. Reorder the array so that all elements with values less than the pivot come before the pivot, while all elements with values greater than the pivot come after it (equal values can go either way). After this partitioning, the pivot is in its final position. This is called the partition operation.
3. Recursively apply the above steps to the sub-array of elements with smaller values and separately to the sub-array of elements with greater values.

We're going to use different techniques to calculate the pivot.

```python
def quicksort(array, pivotIndex = defaultPivot):
    """Return the array passed as argument sorted, using quicksort."""
    if not array:
        return []
    else:
        pivot = array.pop(pivotIndex(array))
        less = [x for x in array if x <  pivot]
        more = [x for x in array if x >= pivot]
        return quicksort(less, pivotIndex) + [pivot] + quicksort(more, pivotIndex)
```

```python
def quicksort2(array, pivotIndex = defaultPivot):
    """Entrance function, to use the quicksort with the second partition strategy, in-place"""
    _quicksort2(array, 0, len(array) - 1, pivotIndex)
 
def _quicksort2(array, first, last, pivotIndex):
    """Recursive method, that uses the second partition strategy for the quicksort"""
    if first < last:
      pivot = partition(array, first, last, pivotIndex)
      _quicksort2(array, first, pivot - 1, pivotIndex)
      _quicksort2(array, pivot + 1, last, pivotIndex)
 
def partition(array, first, last, pivotIndex) :
    """ Partition the array, and compare the elements in order to only have, less than
        the pivot to the left and greater than the pivot to the right of it.
    """
    pivot = pivotIndex(array)
    swap(array, pivot, last)
    for i in range( first, last ):
      if array[i] <= array[last]:
        swap(array, i, first )
        first += 1
 
    swap( array, first, last )
    return first
 
def swap( A, x, y ):
    """Auxiliary function that swap the values in A[x] with A[y], and vice versa."""
    A[x], A[y] = A[y], A[x]
```

```python
def defaultPivot(array):
    """Return the position of the first element of the array."""
    return 0
```

```python
def randomPivot(array):
    """Return a random position of the array"""
    return random.randint(0, len(array) - 1)
```

```python
def medianOfThreePivot(array):
    """Return the position of the median of the first, mid and last element"""
    # (value, index)
    first = (array[0], 0)
    mid = (array[len(array)//2], len(array)//2)
    last = (array[-1], -1)
    # return the index value of the mid element(median). 
    return sorted([first, mid, last])[1][1]
```

We can see by the graph that for exponents bigger than five, quicksort is more efficient than insertion sort. As so let's adapt our algorithm to use insertion sort for arrays with size less than or equal to $2^5$.

```python
def hybridQuicksort(array, pivotIndex = defaultPivot):
    """Return the array passed as argument sorted, using hybrid quicksort."""
    if not array:
        return []
    elif(len(array) <= 2**5):
        return countingSort(array)
    else:
        pivot = array.pop(pivotIndex(array))
        less = [x for x in array if x <  pivot]
        more = [x for x in array if x >= pivot]
        return quicksort(less, pivotIndex) + [pivot] + quicksort(more, pivotIndex)
```

We have to define the different quicksorts, that use different methods to calculate the pivot

```python
# Using the randomPivot method.
def quicksortRand(array):
    """A wrap around the quicksort, to use the randomPivot method"""
    quicksort(array, pivotIndex = randomPivot)
    
def hybridQuicksortRand(array):
    """A wrap around the hybridQuicksort, to use the randomPivot method"""
    return hybridQuicksort(array, pivotIndex = randomPivot)

# Using the median-of-three method to calculate the pivot.
def quicksortMedian(array):
    """A wrap around the quicksort, to use the medianOfThreePivot method"""
    quicksort(array, pivotIndex = medianOfThreePivot)
    
def hybridQuicksortMedian(array):
    """A wrap around the hybridQuicksort, to use the medianOfThreePivot method"""
    return hybridQuicksort(array, pivotIndex = medianOfThreePivot)
```

## Radix Sort

The idea of Radix Sort is to do digit by digit sort starting from least significant digit to most significant digit. Radix sort uses counting sort as a subroutine to sort.

```python
def getDigit(num, base, digit_num):
    """Get the selected digit"""
    return (num // base ** digit_num) % base  
 
def makeBlanks(size):
    """Create a list of empty lists to hold the split by digit, this is going to be used to create the insertion sort"""
    return [[] for _ in xrange(size)]  


def split(array, base, digit_num):
    """Sort elements based on the base and digit_num"""
    buckets = makeBlanks(base)
    for num in array:
        buckets[getDigit(num, base, digit_num)].append(num)  
    return buckets
 
def merge(array):
    """Concatenate the elements back in order"""
    new_list = []
    for sublist in array:
       new_list.extend(sublist)
    return new_list
 
def maxNum(array):
    """largest value element of a list, this is used for the counting sort routine"""
    return max(num for num in array)  
 
def radixSort(array, base = 10):
    """Sort elements in array, using Radix Sort with base = base"""
    passes = int(round(log(maxAbs(array), base)) + 1) 
    new_list = array[:]
    for digit_num in range(passes):
        new_list = merge(split(new_list, base, digit_num))
    return new_list
```

# Median Finding

This group of algorithms are known as selection algorithms, we're going to implement two variations of the median finding algorithms, one of them utilizing the partion by itself (pivot = default) and other utilizing the median-of-median as the pivot.


```python
def medianOfMedians(A, p, r):
    """Find median of the array, based on the medians of subarrays of size 5."""
    # make a copy of the array
    A = A[:]
    
    # divide the n elements of A into n / 5 groups
    groups = [[]] * (((r - p) + 4) / 5)
    for x in range(p, r):
        groups[(x - p) / 5] = groups[(x - p) / 5] + [A[x]]
 
    # find the median of each group
    medians = [sorted(l)[(len(l) - 1) / 2] for l in groups]
 
    # find the median of medians
    if len(medians) == 1:
        median_to_rule_them_all = medians[0]
    else:
        median_to_rule_them_all = select(medians, 0, len(medians), (len(medians) - 1) / 2)
    
    return median_to_rule_them_all

def defaultPivotSelection(A, p, r):
    """Select the first element to be used by the simple partition selection"""
    return p

def partition(A, p, r, x):
    """ partition A[p:r] in place about x, and return the final position of the pivot"""
    # find the index of the pivot
    i = -1
    for j in range(p, r):
        if A[j] == x:
          i = j
          break
 
    # move the pivot to the end
    if i != -1:
        t = A[r - 1]
        A[r - 1] = A[i]
        A[i] = t
 
    # keep track of the end of the "less than" sublist
    store_index = p
 
    for j in range(p, r - 1):
        if A[j] < x:
            t = A[j]
            A[j] = A[store_index]
            A[store_index] = t
            store_index += 1
 
    # put the pivot in its final place
    if i != -1:
        t = A[r - 1]
        A[r - 1] = A[store_index]
        A[store_index] = t
 
    # return the store index
    return store_index
    
# find the ith biggest element in A[p:r]
def select(A, p, r, i, pivot = medianOfMedians):
    x = pivot(A, p, r)
    # partition A around the median of medians
    partition_index = partition(A, p, r, x)
    
    # base case
    if p + i < partition_index:
        return select(A, p, partition_index, i)
    if p + i > partition_index:
        return select(A, partition_index + 1, r, p + i - partition_index - 1)
    return A[p + i]

# wrapper functions to use on the plots.
def selectWithMedian(Array):
    p = 0
    r = len(array)
    i = r // 2
    select(array, p, r, i)
    
def selectWithoutMedian(Array):
    p = 0
    r = len(array)
    i = r // 2
    select(array, p, r, i, pivot = defaultPivotSelection)
```

We will compare the both aproaches to do the selection, for small input arrays and for large arrays. Also comparing the selection algorithms to the sort algorithms, and look how they compare with each other.

```python
# with small arrays
compareAlgorithms([selectWithMedian, selectWithoutMedian], (4, 8))
# with large arrays
compareAlgorithms([selectWithMedian, selectWithoutMedian], (8, 14))

# compared with the sort algorithms

# small inputs
compareAlgorithms([quicksort, quicksort2,hybridQuicksortMedian, radixSort, countingSort, insertionSort, selectWithMedian, selectWithoutMedian], (4, 8))

# large inputs
compareAlgorithms([quicksort,hybridQuicksortMedian, radixSort, countingSort, selectWithMedian, selectWithoutMedian], (8, 14))
# For time issues the insertion sort was removed on the big inputs comparisson.
```

As expected the selection algorithms are considerably better than all the sort algorithms.
